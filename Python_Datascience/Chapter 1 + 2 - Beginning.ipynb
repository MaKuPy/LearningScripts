```python
# Chapter 2: Einführung in Numpy - Part 1
# Daten können in unterschiedlichen Formaten vorliegen (Bilder, Soundfiles, Text etc.)
# Um Sie zu analysieren ist es wichtig, dass Sie in eine Form von Zahlen tranformiert werden
# >> meist werden dafür arrays verwenden (e.g. Bilder als zweidimensionale Arrays mit Pixelintensity)

import numpy as np
# np.__version__ # gibt Version aus

# Python ist eine dynamische Programmiersprache: im Gegensatz zu C oder Java müssen Variablentypen nicht bei der Erstellung
# deklariert werden und eine deklarierte Variable kann einer mit einem anderen Datentyp überschrieben werden
# >> Dies ist möglich, da in Python neben dem Variablenwert auch der Typ bei der Definition festgelegt wird

# X Aufbau von Python Integers
# tatsächlich enthalten Python Intergers (nach C-Source Code) vier Informationen
# - ob_refcnt: Reference count, die Python hilft die Variable zum Speicher zuzuordnen
# - ob_type: enthält den Typ der Variable
# - ob_size: enthält die Größe der Daten
# - ob_digit: enthält den eigentliche Wert

# Da Listen in Python aus verschiedenen Datentypen bestehen können, muss jedes Item die oben genannten Informationen enthalten
# >> NumPy_Arrays legen den Datentyp fest, sind daher weniger flexibel aber deutlich effizienten


# X Fixed-type arrays in Python
# - Built-in array
from array import array
L = list(range(10))
A = array('i', L) # erstellt einen integer('i')- Array aus der Liste
print(A)

# >> Der built-in array speichert Daten zwar in effizienter Weise ab, aber der np.array bietet nützliche Array-Operationen

# - np Array
print(np.array([1, 4, 2, 5, 3]))

# NumpyArrays benötigen die selbe Art von Objekten
print(np.array([3.14, 4, 2, 3])) # konvertiert den Inhalt in floats
# Alternativ kann der Typ mit dtype bestimmt werden

print(np.array([1, 4, 2, 5, 3], dtype='float32')) # auch float32 ohne '' möglich

# Numpy Arrays können problemlos multidimensional sein
print("\n multidimensionale Array")
print(np.array([range(i, i+3) for i in [2, 4, 6]]))

# X Arrays aus dem nichts erstellen
print("\n Next Part")

print(" Nullen")
print(np.zeros(10, dtype=int))# Erstellt array mit 10 Nullen

print("\n Mehrdimensional")
print(np.ones((3,5), dtype = int)) # Reihe und Spaltenzahl, kann als Tuple mitgegeben werden

print("\n Bestimmter Wert")
print(np.full((2,8), 23)) # Reihen kännen auch mit einem bestimmten Wert gefüllt werden

print("\n lineare Sequenz")
print(np.arange(10, 15, 0.5)) # wie range(start, ende, schritte)

print("\n 5 Werte in Bereich 0-1")
print(np.linspace(0, 1, 5))

print("\n Zufallswerte")
print(np.random.random((3,3)))
print(np.random.normal(0, 1, (3,3))) # Auch normalverteilt mit (M=0, SD=1)


print("\n Identity Matrix")
print(np.eye(4)) # Diagonale 1 rest 0

# Datentypen in Numpy
# e.g. bool_, int_, int32, int64, uint64 (ohne - Werte), float32, float64 etc.
```

```python
# Chapter 2: Einführung in Numpy - Part 2
# X Numpy Array Attributes (size, shape, memory, type)
import numpy as np
np.random.seed(0)

x1 = np.random.randint(10, size=6)
x2 = np.random.randint(10, size=(3, 4)) # Reihen, Spalten
x3 = np.random.randint(10, size=(3, 4, 5)) # Tabellen, Reihen, Spalten

print("Attribute")
print(f"x3 ndim: {x3.nbytes}, x3 shape: {x3.shape}, x3 size: {x3.size}, x3 type: = {x3.dtype}")
print(f"Itemsize = {x3.itemsize}, nbytes = {x3.nbytes}")
# itemsize * size = nbytes

# X Array Indexing
print("\n Array Indexing \n Eindimensionale Arrays")
# Ein Element aus x1 kann mit einem einzelnen Index ausgewählt werden
print(x1[0], x1[[0, 4]]) # oder mehre mit einer Liste

# In zweidimensionalen Arrays (x2) wird ein comma-seperated Tuple mit Reihe, Spalte benötigt
print("\n Zweidimensionale Arrays")
print(x2)
print(x2[0,0], x2[2, 0])

# so können auch einzelne Werte geändert werden
x2[0, 0] = 13
print(x2)

# Mit der x[start:ende:step] Konnotation können ganze Abschnitte von Werten ausgewöhlt werden
print("\n mehrere Werte auswählen")
x = np.arange(10)
print(x[:5]) # fehlende Angaben = 0
print(x[4:7])
print(x[::2]) # jedes zweite Element
print(x[::-1]) # jedes Element rückwärts

# Die gleiche Logik funktioniert für Mehrdimensionale Arrays
print(x2[0, :]) # wählt komplette erste Reihe
# Alternative Reihenauswahl
print(x2[0])
print(x2[1:2, :3]) # mittlere Reihe, 1st 3 Spalten
print(x2[::-1, ::-1]) # dreht Array um

# Subarrays return views rather than copies
print("\n Arrays und Kopien")
print(x2)
x2_sub = x2[:2, :2]
print(x2_sub)
print("Veränderungen ändern das Orignial")
x2_sub[0, 0] = 99
print(x2)
print(x2_sub)
# >> dies kann hilfreich sein, da mit Teilen des Datensatzes gearbeitet werden kann, als wäre es der tatsächliche Datensatz

# Kopien erzeugen, die nicht verändert werden
print("\n Kopien bewusst erzeugen")
x2_sub_copy = x2[:2, :2].copy()
print(x2_sub_copy)
x2_sub_copy[0,0] = 14
print(x2_sub_copy)
print(x2)

# X Reshaping arrays
print("\n Reshaping")
# Um einen array in eine anderes Format zu bringen kann die reshape-Methode verwendet werden
print(np.arange(1,10).reshape((3,3)))

# Ein eindimensionaler Array kann so auch in einem zweidimensionalen verwandelt werdne
x = np.array([1, 2, 3])
print(x.reshape((1, 3)))
print(x.reshape((3, 1)))
# Alternativ kann das np.newaxis keyword verwendet werden
print(x[np.newaxis, :])
print(x[:, np.newaxis])

# Arrays erweitern
print("\n Arrays erweitern")
print(np.concatenate([x, x])) # mit concatenate können zwei Arrays einfach aufaddiert werden

# Funktioniert auch mit zweidimensionalen arrays
grid = np.array([[1, 2, 3], 
                [4, 5, 6]])
print(np.concatenate([grid, grid]))
# Auch in der Reihe
print(np.concatenate([grid, grid], axis = 1))
# Alternativ kann vstack oder hstack verwendet werden
print(np.vstack([grid, grid]))
print(np.hstack([grid, grid]))

# Arrays zerteilen
print("\n Arrays zerlegen")
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5]) # Liste von Index gibt Anfangsindex des neuen Abschnittes an
print(x1, x2, x3)

# genauso kann mit vsplit und hsplit ein Array anhand des Spalten- bzw. Reihenindexes zerlegt werden
grid = np.arange(16).reshape((4, 4))
print(grid)

x1, x2 = np.vsplit(grid, [2])
print(x1)
print(x2)
```

```python
# Chapter 2: Einführung in NumPy - Part 3
# Universal Functions (ufuncs) in NumPy erlauben schnelle Rechenoperationen mit Arrays

# Wiederholte Operationen mit Loops sind in Python aufgrund der dynamischen Typen sehr langsam
# e.g. Kehrwert einer Liste
np.random.seed(0)
def berrechne_kehrwert(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0/values[i]
    return output

# values = np.random.randint(1, 100, size=1000000)

# %timeit berrechne_kehrwert(values) # braucht ca. 1.84 Sekunden >> Sehr lang für einfachen Code
# Das Problem ist dabei, das Python zum speichern und ausführen jedes mal den Typ der Variable prüfen und abspeichern muss

# Berechnung mit ufuncs
# Bei NumPy werden Berechnungen auf jedes Element des Arrays angewandt (vectorization, ufuncs)
# %timeit 1/values # benötigt 2.88 ms im Schnitt > deutlich schneller als die Loop-Vorgehensweise

# Weitere Beispiele
print(np.arange(5) / np.arange(1, 6))
x = np.arange(9).reshape((3,3))
print(2 ** x) # 2 hoch Array

# X 2 Typen von ufuncs
# Unary ufuncs (ein Input)  und binary ufuncs (two Inputs)
# Einfache Arithmetische Operationen 
print("\n Arithmetische Operationen")
x = np.arange(4)
print(x)
print(x + 3, "plus 3")
print(x * 3, "mal 3")
print(x // 3, "Ganzzahlige Div 3")
print(x % 3, "Modulo Div 3")
print(x - 3, "minus 3")
print(x / 3, "Div 3")
# alternativ np.add(), np.subtract(), np.negative(), np.multiply(), np.divide(, np.floor_divide(), np.power(), np.mod()

# Weitere Funktionen
print("\n Weitere Funktionen")
# absolute Werte
print("abs", x-2, np.abs(x-2)) # alternativ np.absolute() >> ufunc von abs()

# X Trigonometrische Funktionen
theta = np.linspace(0, np.pi, 3) # generiert Winkel?
print("theta", theta)
print("sin(theta)", np.sin(theta))
print("cos(theta)", np.cos(theta))
print("tan(theta)", np.tan(theta))
# mit np.arcsin(), np.arccos() und np.arctan() sind auch die Inversen trigonometrischen Funktionen verfügbar

# X Exponential Funktion und Logarithmus
x = np.array([1, 2, 3, 10])
print("exp", np.exp(x))
print("^3", np.power(x, 3))
print("ln", np.log(x))
print("log10", np.log10(x))

# >> Weitere Funktionen im scipy.special MOdul (from scipy import special) >> check doc

# X Speicherort angeben mit ufucs
print("\n Seperate Objecte speichern")
x = np.arange(5)
y = np.zeros((2, 5))
print(y)
np.multiply(x, 10, out=y[0,:]) # Kann auch in Zeile oder Spalte gespeichert werden
print(y)

# X Arregieren mit reduce
print("\n Aggregieren")
x = np.arange(1, 6)
print("sum", x, np.add.reduce(x))
print("mean", np.add.reduce(x)/len(x))
# auch mit multiply, subtract etc. möglich

# Zwischenergebnisse mit aggregate speichern
print("cumSum", x, np.add.accumulate(x))

# X Kombinieren
# mit outer wird jedes Element zweier Vektoren miteinander kombiniert
y = np.arange(1,12,2)
print("Outer-Kombination")
print(x)
print(y)
print(np.add.outer(x, y))


# X Min, Max etc.
print("\n Summary Stats")
x = np.arange(0, 6).reshape((2, 3))
print("arraySUM", np.sum(x))
print("min", np.min(x[1,:])) # Min in Zeile 1 (zweite Zeile, da Start 0)
print("max", np.max(x))
print("mean", np.mean(x))
print("std", np.std(x))
# Alternativ kann x.max() oder x.sum() aufgerufen werden
# Dabei kann eine Achse gewählt werden für die die Operation berechnet werden soll
print("min per col", x.min(axis=0))

# Weitere Aggregierende Funktionen: 
# np.nansum(), np.nanprod(), np.nanmean(), np.nanstd(), np.nanmin(), np.nanmax(), np.nanargmin(>findet Index von min),
# np.namargmax(), np.nanmedian(), np.nanpercentile() etc.

# Beispiel: Tabelle mit Presidenten-Größen
print("\nAnalysing President Heights")
import pandas as pd
dat = pd.read_csv("president_heights.csv")
# in Numpy array umwandeln
heights = np.array(dat['height(cm)'])
print("mean height", np.nanmean(heights))
print("std height", np.nanstd(heights))

print("min height", np.nanmin(heights), "cm,", dat['name'][np.nanargmin(heights)])
# Quartile
print("1st Quartile", np.percentile(heights, 25))
print("Median", np.nanmedian(heights))
print("3rd Quartile", np.percentile(heights, 75))
print("max height", np.nanmax(heights),"cm,", dat['name'][np.nanargmax(heights)])

# Kann mit matplotlib visualisiert werden
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # set plot style
plt.hist(heights)
plt.title('Height Distribution for US Presidents')
plt.xlabel('Height (cm)')
plt.ylabel('n´Number')
```

```python
# Chapter 2: Einführung in NumPy - Part 4
# X Broadcasting ...
# ... wird verwendet, um binary ufuncs auf Arrays von verschiedenen Größen anzuwenden
# Dabei wird der kleinere Array, so erweitert, dass er auf den Größeren passt
print("X Broadcasting")
x = np.array([1, 2, 3])
print(x + 5) # entspricht [1, 2, 3] + [5, 5, 5]
y = np.ones((3,3))
print(x+y) # der Array [1, 2, 3] wird über drei Reihe gestreckt/gebroadcasted

# Broadcasting, kann auch zu einer Dimensionsvergrößerung des Arrays führen
print("\n komplett unterschiedliche Dimensionen")
y = x.copy()[:,np.newaxis]
print(x + y) # Beide Arrays werden in die jeweils andere Dimension gestreckt

# X Eine Variable mit broadcasten zentrieren
print("\n e.g. Zentrieren")
x = np.random.random((10, 3))

xmean = x.mean(axis=0) # Berechnet colmeans 
print(xmean)
x -= xmean
print(x.mean(0)) # Werte gehen gegen 0

print("\n e.g. two-dim Funktion Visualisieren")
x = np.linspace(0, 5, 50)
y = x.copy()[:, np.newaxis]

z = np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x) 
#plt.imshow(z, origin ='lower', extent =[0, 5, 0, 5],
#          cmap='viridis')
#plt.colorbar()


# X Boolean Masking
# Masking wird dann relevant, wenn man Daten auf Basis von einem Auswahlkriterium darstellen, zählen, zusammenfassen möchte
print("\n Boolean Masking")

rainfall = pd.read_csv('Seattle2014.csv')['PRCP'].values # importiert nur Regenvariable
inches = rainfall/ 254
print(inches.shape) # Enthält 365 Regenwerte, einer für jeden Tag im Jahr
plt.hist(inches, 40)

# ufuncs ermöglichen Elementweise Vergleiche von Werten
x = np.arange(6)
print(np.greater_equal(x, 3)) # Das Ergebnis ist ein array von Booleans

# ufuncs: np.equal(), np.not_equal(), np.less(), np.less_equal(), np.greater(), np.greater_equal()

# Operationen mit Boolean-Arrays
x = np.random.randint(1, 10, 15).reshape((3,5))
print(x)
print("Wie viele Werte sind größer 6?")
print(np.count_nonzero(x > 6), "Werte größer 6") # np.sum() würde zu dem gleichen Ergebnis führen
# np.sum(x > 6, axis = 1)# , würde für jede Reihe die Summe gebildet werden

# mit np.any() kann überprüft werden, ob mindest ein Wert dem Kriterium entspricht
# mit np.all() wird geprüft, ob alle Werte dem Kriterium entsprechen

print(np.sum(np.greater(inches, .5) & np.less(inches, 1)), "Tage mit mehr als 0.5 inch Regen und weniger als 1 inch")
# Ergebnisse
print("results")
print("Tage ohne Regen:", np.sum(np.equal(inches, 0)))
print("Tage mit Regen:", np.sum(np.greater(inches, 0)))

# Auswahl mithilfe von Boolean arrays
rainy = inches[np.not_equal(inches, 0)] # Wählt alle Tage aus, an denen es geregnet hat
print("Median:", np.round(np.nanmedian(rainy), 3))
print("Mean:", np.round(np.nanmean(rainy), 3))


# X Fancy Indexing
print("\nFancy Indexing")
x = np.random.randint(100, size=10)
# Mehrere ELemente können auf zwei Arten ausgewählt werden
print([x[3], x[5], x[7]])
# oder mit einer Liste von Werten = Fancy Indexing
print(x[[3, 5, 7]])

# Über den Indexarray wird die Form des Outputs bestimmt
index = np.array([[3, 5],
                  [7, 9]])
print(x[index])

# Fancy Indexing funktioniert auch in mehreren Dimensionen
x = np.arange(12).reshape(3,4)
row = np.array([0, 1, 2])
col = np.array([2, 1, 3])
print(x[row, col], "Multidim Choice")

# Damit können auch Werte modifiziert werden
x[row, col] = -99
print(x)

# Sorting Arrays
print("\n X Sorting")
# Es gibt verschiedene Sorting Algorithmen, um einen Array zu sortieren:
# - insertion sort (speichert einen Key-Element und vergleicht es mit dem vorherigen Element der Liste und tauscht Positionen)
# - selection sort (das Minimum des unsortierten Teils wird mit dem Wert beim Index der aktuellen Iteration getauscht)
# - merge sort (Zerteilt den Datensatz in seperate sub-arrays, die dann sortiert werden)
# - quick sort (Wählt einen zentralen Wert als Vergleichselement (e.g. Median) und platziert Werte, die kleiner/größer sind davor/-hinter)
# - bubble sort (Vergleicht stets nebeneinanderliegende Elemente und tauscht deren Position)
# - bogosort (shuffled den Datensatz so lange bis er in richtiger Reihenfolge vorliegt)

# np.sort() ist ein Algorithmus in numpy, der quicksort verwendet!
X = np.array([2, 1, 4, 3, 5])
print("mit np.sort()", np.sort(X)) # verändert den Array nicht
print(X) 
X.sort() # verändert den Array
print("mit.sort()", X)

# argsort() gibt die Indexe des sortierten Arrays zurück
X = np.array([2, 1, 4, 3, 5])
print("Indexe", np.argsort(X))

# Über unterschiedliche Dimensionen sortieren
rand = np.random.RandomState(42)
X = np.random.randint(0, 10, (4, 6))
print("\n", X)

# Jede Spalte sortieren
print("\n Spalten sortieren\n", np.sort(X, axis = 0))

# Jede Reihe sortieren
print("\n Reihen sortieren\n", np.sort(X, axis = 1))

# um die k-kleinesten Werte zu finden, wird die Partition Funktion gewählt
print("\n Partitioning - 2 kleinste Werte in Reihe\n", np.partition(X, 2, axis=1))

# Zwei nahsten Punkte finden
print("\n Nahe Punkte finden")
X = np.random.rand(10, 2)
print(X)

# Visualisierung zum verdeutlichen
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
plt.scatter(X[:, 0], X[:, 1], s=100)

# Distanz berechnen
print(X[:, np.newaxis,:])
print(X[np.newaxis, :, :])
dist_sq = np.sum((X[:, np.newaxis,:] - X[np.newaxis, :, :])**2, axis=-1)
print("\n Distanzmatrix\n", dist_sq)

nearest_partition = np.argpartition(dist_sq, 2+1, axis = 1)
plt.scatter(X[:,0], X[:,1], s=100)
# Linien ergänzen
for i in range (X.shape[0]):
    for j in nearest_partition[i, :2+1]:
        plt.plot(*zip(X[j], X[i]), color="blue")
```

```python
# Structured Arrays
# NumPys structured und record arrays sind eine gute Möglichkeit, um Arrays zu speichern, die unterschiedliche Datentypen enthalten :D
# Oft werden jedoch stattdessen Pandays DataFrame-Objekte verwendet

name = ['Alice', 'Knorke', 'Drouga', 'Lupin']
age = [87, 12, 56, 44]
weight = [55.0, 34.8, 85.5, 61.5]

# mit np.zeroes kann über ein Diktionär jeder Spalte ein Name gegeben werden und diese einem Format zugeordnet werden
data = np.zeros(4, dtype={'names': ('name', 'age', 'weight'),
                          'formats':('U10', 'i4', 'f8')})
# nun kann der Datensatz gefüllt werde
data['name'] = name
data['age'] = age
data['weight'] = weight
print(data)

# Auswahl über Spaltennamen
print(data['name'])
# oder Zeilenindex
print(data[0])

# oder Kombination mit [][]
print(data[-1]['name']) # Gibt Name der letzten Person 
# print(data[-1, 0]) funktioniert nicht
print(data[data['age'] < 50]['name']) # gibt Name der Personen, deren Alter < 50 ist

# RecordArrays funktionieren auf eine ähnliche Weise, nehmen jedoch etwas mehr Speicher ein, da sie mit .varname Spalten aufrufen können
data_rec = data.view(np.recarray)
print(data_rec.age)
```

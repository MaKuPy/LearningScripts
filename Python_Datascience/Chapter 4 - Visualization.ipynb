```python
# XX Data Science Handbook
#  X Kapitel 4: Visualisieren mit Matplotlib

# Matplotlib: eine der verbreitesten Librarys für Datenvisualisierung
# >> neuere Pakete, wie ggplot, Seaborn, HoloViews, Altair oder Pandas können die Visualisierung verbessern

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np


# Styles
# ... können via plt.style.use('style') gesetzt werden
plt.style.use('classic') # stellt classic style ein

# Plotting from Script
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show() # Gibt den PLot im Terminal aus
# Sollte nur einmal am Ende des Skripts verwendet werden
# >> Da diese Funktion komplex mit dem Backend interagiert, was bei mehrfacher Ausführung zu Problemen führen kann
```

```python
# %matplotlib notebook # öffnet Interaktive Plots mit plot.show() in seperatem Fenster

%matplotlib inline
# gibt statische Plots im Notebooks aus, sobal der plt.plot() Befehl ausgeführt wird

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

# Datei speichern
fig.savefig("my_figure.png") # .jpg, .jpeg, .svg etc auch möglich
```

```python
# Plotten mit zwei Interfaces
# X MATLAB-Style

plt.figure() # erstellt ein figure-Objekt
# Erstellt das erste Panel und legt Achsen fest
# "Grid mit zwei Reihen einer Spalte" >> Erster Platz (Panel1 Oben)
plt.subplot(2, 1, 1) # reihe, spalte, panel (subplot(2,2,1) würde ein 2x2 Grid mit 4 Feldern erstellen)
plt.plot(x, np.sin(x))

# "Füllt auf dem debestehenden Grid den zweiten Platz (Panel2 unten)"
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# X Obejktorientiertes Interface
fig, ax = plt.subplots(2)
# Erstellt ein Interface ax mit 2 Plätzen ax[0] und ax[1], die mit.plot gefüllt werden können

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# X Simple Line Plots

plt.style.use('seaborn-whitegrid') # Styletamplete mit Rastern
fig = plt.figure() # Erstellt eine Grafik (Container für Achsen, Label, Text, Graphen etc.)
ax = plt.axes() # Erstellt eine Achse (= Koordinatensystem)

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x), label='sin(x)') # plt.plot() würde ebenfalls funktionieren
# jede Funktion kan geplottet werden, z.B. auch (x, x*x) oder (x) 

# Diese werden dann über den aktuellen Plot gelegt

# X Grafische Anpassung
# - Farben und Linientyp festlegen
ax.plot(x, np.sin(x-1), color='g', linestyle='solid', label='sin(x-1)') # kurzbezeichnung
ax.plot(x, np.sin(x-2), color='#FFDD44', linestyle='dashed', label='sin(x-2)') # Hexcode
ax.plot(x, np.sin(x-3), color=(1.0, 0.2, 0.3), linestyle='dashdot') # rgb-tuple mit WErten zwischen 0 und 1
ax.plot(x, np.sin(x-4), color='purple', linestyle="dotted") # HTML-Bezeichnungen

# alternative kann linestyle auch mit ='-', '--', '-.', '..' festgekegt werden


# - Achsenbegrenzung festlegen
# ... mit .xlim und .ylim
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)

# ... alternative mit plt.axis()
# plt.axis([xmin, xmax, ymin, ymax]) kann sämtliche Achsen festlegen
plt.axis([-1, 11, -1.5, 1.5])
# plt.axis('tight') würde die Achsen so festlegen, dass sie direkt am Plot dran sind


# - Plotlabel
# ... mit plt.title(), plt.axlabel(), plt.ylabel()

plt.title("Eine Sinus-Kurve", fontsize=25) # fontsize ist optional
plt.xlabel('x')
plt.ylabel('sin(x)')

# >> siehe weitere optionale Argument: 
# https://matplotlib.org/stable/gallery/text_labels_and_annotations/accented_text.html#sphx-glr-gallery-text-labels-and-annotations-accented-text-py

# -Legende
# ... wird mit plt.legend() automatisch erstellt, wenn bei plot, das optionale Argument label='name' gesetzt wurde
plt.legend(fontsize=12, frameon=True, loc='upper left')
# ... viele weitere Argumente für Legenden (e.g. fancybox=True, framealpha=1, borderpad=1, shadow=True etc)
```

```python
# X Scatterplots
x = np.linspace(0, 10, 30)
y = np.sin(x)

# ... mit plot und 'o'
#rng = np.random.RandomState(0)
fig = plt.figure(figsize = (11, 7))
plt.subplot(2, 2, 1)
plt.plot(x, y, 'o', color='blue')
# >> 'o' gibt die Art der Punkte an (siehe unten)

plt.subplot(2, 2, 2)
plt.plot(x, y, '-p', color='blue',
         markerfacecolor='yellow',
         markeredgewidth = 2,
         markeredgecolor = 'red',
         markersize=12, linewidth=3)

# >> mit '-o' wird außerdem eine Linie hinzugefügt
# >> viele weitere Anpassungen möglich

# ... mit .scatter()
plt.subplot(2, 2, 3)
plt.scatter(x, y, marker='o')

# der Vorteil von scatter ist, dass dabei grapfische Argumente mit Variablen belegt werden können
plt.subplot(2, 2, 4)
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = abs(1000 * rng.randn(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

```python
# X Errorbars 
fig = plt.figure(figsize=(5, 5))
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.subplot(2, 1, 1)
plt.errorbar(x, y, yerr=dy, fmt='.k')

# Graphische Gestaltung
plt.subplot(2, 1, 2)

plt.errorbar(x, y, yerr=dy, fmt='o', color='blue',
            ecolor='royalblue', elinewidth=3, capsize=0)
# fmt - bestimmt Art des Punktes
# color - die Farbe des Punktes
# ecolor - Die Farbe der Errorlinie, elinewidth - deren Breite
# capsize - die Größe der Linie an den Enden

# >> es können auch kontinuierliche Fehlerterme mit der Gaussian process regression abgebildet werden (wenn relevant nachlesen)
```

```python
# X Desity/Contourplots
fig = plt.figure(figsize=(4, 3))

# visualisieren drei Dimensionen zweidimensional
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10+ y * x) *np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z, 20, cmap='RdGy')

# Wenn relevant nachlesen
```
